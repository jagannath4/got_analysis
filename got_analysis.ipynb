```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import warnings
warnings.filterwarnings('ignore')
```

```python
df = pd.read_csv('/home/kali/Downloads/Game_of_Thrones.csv')
```

```python
plt.figure(figsize=(30,30))
for i in range(8):
    data = df.copy()
    data = data[data['Season'] == i+1]
    plt.subplot(4,2,i+1)
    sns.barplot( x=data['No. of Episode (Season)'] , y=data['IMDb Rating']) 
    plt.ylabel('No. of Episode', fontsize=15)
    plt.xlabel('IMDB Rating', fontsize=15)
    plt.title("Season" + str(i+1), fontsize=15)
plt.show()

```
